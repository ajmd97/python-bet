```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
url = input('Introduce la url a scrapear: ')
besoccer = requests.get(url)

# por ejemplo: https://es.besoccer.com/livescore/2022-09-17
```

```python
besoccer.status_code
```

```python
bs_soup = BeautifulSoup(besoccer.text, 'lxml')
```

```python
type(bs_soup)
```

```python
print(bs_soup.prettify())
```

```python
links = bs_soup.find('div', attrs={'class':'matches'}).select('a[href*="/partido/"]')
links
```

```python
link = links[0]
link
```

```python
link_match = [link.get('href') for link in links]
link_match
```
